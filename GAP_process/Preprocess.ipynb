# Summary GAP patient and slide data

```python
import os
import numpy as np
import pandas as pd
```

```python
path = '/home/zhong/data/GAP/'
label_path = path + 'Label/'
wsi_path = path + 'WSI/'
```

```python
dirlist = os.listdir(label_path)
dirlist = sorted(dirlist)
len(dirlist)
```

```python
patients = [] 
for i, name in enumerate(dirlist): 
    patients.append(name.split('-')[0])
    dirlist[i] = name.strip('_L.jpg')
num = []
tags = sorted(list(set(patients)))
for tag in tags: 
    num.append(patients.count(tag))
GAP_info = pd.DataFrame({
    'tag': tags,
    'wsi_num': num
})
GAP_info
```

```python
GAP_info['file'] = [[] for _ in range(len(GAP_info))]
begin = 0
for i, n in enumerate(num):
    GAP_info['file'][i] = dirlist[begin:begin+n]
    begin+=n

GAP_info
    
```

```python
GAP_info.to_csv(path + 'GAP_info.csv')
```

# Summarize patch information

```python
import pandas as pd
import os
```

```python
path = '/home/zhong/data/GAP/'
patch_path = path + 'Patch_224/'
```

```python
GAP_info = pd.read_csv(path + 'GAP_info.csv')
GAP_info
```

```python
patch_info = pd.DataFrame({
    'name': GAP_info['tag']
})
types = ['CTne','IT','LE','CT','CTmvp','CTpan','CTpnz','BG']
for type in types:
    patch_info[type] = [0 for i in range(len(patch_info))]
for i in range(len(patch_info)):
    tag_path = patch_path + patch_info['name'][i] + '/'
    for type in types:
        patch_info[type][i] = len(os.listdir(tag_path + type))
patch_info.to_csv('patch_224_info.csv')
```

# Construction of train set and validation set

```python
import pandas as pd
import os
import numpy as np
import shutil
import random
from tqdm import tqdm
import cv2
```

```python
path = '/home/zhong/data/GAP/'
patch_path = path + 'Patch_224/'
data_path = path + 'Dataset_224_1024/'
```

```python
info = pd.read_csv(path + 'patch_info_sum.csv')
types = list(info)[-8:]
train_val = pd.DataFrame({
    'name': info['name'],
    'usage': ['train' for i in range(len(info))]
})
train_val['usage'][29] = 'val'
train_val['usage'][30] = 'val'
train_val['usage'][31] = 'val'
train_val['usage'][20] = 'val'
train_val['usage'][23] = 'val'
```

```python
sample_num = {
    'train': 1024,
    'val': 100
}
for state in ['train','val']:
    print(f'===Begin to select {state} set===')
    for type in types: 
        print(f'---Begin to process {type}---')
        type_path_data = data_path + state + '/' + type
        if os.path.exists(type_path_data) is not True: 
            os.mkdir(type_path_data)
        name_list = list(train_val[train_val['usage']==state]['name'])
        dir_list = []
        for name in name_list:
            type_path_patch = patch_path + name + '/' + type + '/'
            dir_list += [type_path_patch + dir for dir in os.listdir(type_path_patch)]
        print(f'{len(dir_list)} of {type} patches')
        dir_list_sample = random.sample(dir_list, sample_num[state])
        print('Begin to copy')
        for dir in tqdm(dir_list_sample):
            shutil.copy(dir, type_path_data)
        print('Finished!')
    print(f'*****{state} set finished*****')    
        
```
